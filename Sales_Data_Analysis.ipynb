```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

df = pd.read_csv("sales_data_sample.csv", encoding='latin1')
df.head()
```

```python
df = pd.read_csv("sales_data_sample.csv", encoding='ISO-8859-1')
df.head()
```

```python
df.isnull().sum()
df = df.dropna()
df = df.dropna()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['Month'] = df['ORDERDATE'].dt.month
```

```python
monthly_sales = df.groupby('Month')['SALES'].sum()

plt.figure()
monthly_sales.plot(kind='line')
plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.show()
```

```python
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum()

plt.figure()
product_sales.plot(kind='bar')
plt.title("Sales by Product Line")
plt.show()
```

```python
country_sales = df.groupby('COUNTRY')['SALES'].sum().head(10)

plt.figure()
country_sales.plot(kind='bar')
plt.title("Top 10 Countries by Sales")
plt.show()
```

Insights:
1. Sales show seasonal trends with peaks in certain months.
2. Classic Cars and Vintage Cars are the best-selling product lines.
3. The USA and France contribute the highest sales.
